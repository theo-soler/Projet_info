# Web scrapping boursorama


```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.boursorama.com/bourse/matieres-premieres/cours/historique/_HR'
```

```python
r = requests.get(url)
```

```python
soup = BeautifulSoup(r.content, 'html.parser')
```

```python
soup
```

```python
rows = soup.select('tbody tr td')
```

```python
rows
```
